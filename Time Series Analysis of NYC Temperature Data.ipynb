**ARIMA, SARIMA AND PROPHET MODEL FOR TRAINING DATA**

```python
# Import libraries
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from prophet import Prophet
```

```python
# 1. Split data
train = nyc_data['1970':'2003']  # ~408 months
test = nyc_data['2004':'2013']   # ~120 months
print(train.shape, test.shape)
```

```python
# 2. ARIMA: Check stationarity
adf_result = adfuller(train['AverageTemperature'])
print(f"ADF p-value: {adf_result[1]:.3f}")  # If >0.05, difference data
```

```python
# ARIMA model (try p=1, d=1, q=1)
arima_model = ARIMA(train['AverageTemperature'], order=(1, 1, 1))
arima_fit = arima_model.fit()
arima_forecast = arima_fit.forecast(steps=120)
```

```python
# SARIMA model (try seasonal order=(1,1,1,12))
sarima_model = ARIMA(train['AverageTemperature'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_fit = sarima_model.fit()
sarima_forecast = sarima_fit.forecast(steps=120)
```

```python
# Prophet: Prepare data
prophet_df = train.reset_index().rename(columns={'date': 'ds', 'AverageTemperature': 'y'})
prophet_model = Prophet(yearly_seasonality=True)
prophet_model.fit(prophet_df)
future = prophet_model.make_future_dataframe(periods=120, freq='MS')
prophet_forecast = prophet_model.predict(future)
prophet_yhat = prophet_forecast['yhat'][-120:]  # Last 120 months
```

```python
prophet_tail = prophet_forecast.tail(120).set_index('ds')['yhat'] 
```

**FORECASTS ON MODELS FOR TEST DATA**

```python
# Save forecasts
forecasts = pd.DataFrame({
    'ARIMA': arima_forecast,
    'SARIMA': sarima_forecast,
    'Prophet': prophet_tail
}, index=test.index)
forecasts.to_csv('forecasts_test.csv')
```

```python
# Import metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Load forecasts
forecasts = pd.read_csv('forecasts_test.csv', index_col=0, parse_dates=True)
actual = test['AverageTemperature']
```

```python
# Load forecasts
forecasts = pd.read_csv('forecasts_test.csv', index_col=0, parse_dates=True)
actual = test['AverageTemperature']
```

**EVALUATION METRICS (MAE, RMSE, MAPE)**

```python
# Compute metrics
metrics = {}
for model in ['ARIMA', 'SARIMA', 'Prophet']:
    mae = mean_absolute_error(actual, forecasts[model])
    rmse = np.sqrt(mean_squared_error(actual, forecasts[model]))
    mape = 100 * np.mean(np.abs((actual - forecasts[model]) / actual))
    metrics[model] = {'MAE': mae, 'RMSE': rmse, 'MAPE': mape}
metrics_df = pd.DataFrame(metrics).T
print(metrics_df)
```

**PLOTS**

```python
# Plot forecasts
plt.figure(figsize=(12, 6))
plt.plot(train['AverageTemperature'], label='Train', alpha=0.5)
plt.plot(test['AverageTemperature'], label='Actual', color='blue')
for model in ['ARIMA', 'SARIMA', 'Prophet']:
    plt.plot(forecasts[model], label=f'{model} Forecast', linestyle='--')
plt.title('Actual vs. Forecasted Temperatures (2004–2013)')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.savefig('forecast_comparison.png')
plt.show()
```

```python
# Select best model
best_model = metrics_df['RMSE'].idxmin()
print(f"Best model: {best_model}")
```

```python
# Forecast 240 months
future_forecast = sarima_fit.forecast(steps=240)
future_index = pd.date_range(start='2014-01-01', end='2033-12-01', freq='MS')
future_df = pd.DataFrame({'Forecast': future_forecast}, index=future_index)
```

```python
# Plot historical and forecast
plt.figure(figsize=(12, 6))
plt.plot(nyc_data['AverageTemperature'], label='Historical', color='blue')
plt.plot(future_df['Forecast'], label='Forecast', color='red', linestyle='--')
plt.title('NYC Temperature Forecast (2014–2033)')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.savefig('future_forecast.png')
plt.show()
```

```python
# Estimate rate of increase
hist_mean = nyc_data['AverageTemperature'].mean()
forecast_mean = future_df['Forecast'].mean()
rate = (forecast_mean - hist_mean) / 2  # °C per decade
print(f"Rate of increase: {rate:.2f}°C/decade")
```

```python
# Historical and forecast means
hist_mean = nyc_data['AverageTemperature'].mean()
forecast_mean = future_df['Forecast'].mean()
rate = (forecast_mean - hist_mean) / 2  # °C/decade
print(f"Historical mean: {hist_mean:.2f}°C, Forecast mean: {forecast_mean:.2f}°C")
print(f"Rate: {rate:.2f}°C/decade")
```

```python
# Seasonal comparison
hist_monthly = nyc_data.groupby(nyc_data.index.month)['AverageTemperature'].mean()
forecast_monthly = future_df.groupby(future_df.index.month)['Forecast'].mean()
seasonal_diff = forecast_monthly - hist_monthly
plt.figure(figsize=(8, 4))
plt.plot(seasonal_diff, marker='o')
plt.title('Seasonal Temperature Change (2014–2033 vs. 1970–2013)')
plt.xlabel('Month')
plt.ylabel('Temperature Difference (°C)')
plt.savefig('seasonal_diff.png')
plt.show()
```
